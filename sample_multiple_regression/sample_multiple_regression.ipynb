```python
import pathlib

import pandas as pd
import matplotlib.pyplot as plt 

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

#データセットの読み込み

```python
dataset_path = "C:\\Users\\over-\\python\\01_keras\\02_回帰\\sample_multiple_regression\\sample_test_data.csv"
```

```python
raw_df_dataset = pd.read_csv(dataset_path)

dataset = raw_df_dataset.copy()
dataset.head()
```

#データクレンジング

```python
print(dataset.isna().sum())
dataset = dataset.dropna()
```

#データの分割

```python
train_dataset = dataset.sample(frac=0.9, random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

```python
train_stats = train_dataset.describe()
train_stats.pop("ave_devi")
train_stats = train_stats.transpose()
print(train_stats)
```

```python
name = train_dataset.columns[30]
print(name)

train_labels = train_dataset.pop(name)
test_labels = test_dataset.pop(name)
print(train_labels)
```

#データの正規化

```python
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
normed_train_data.describe()
```

#モデル作成

```python
def build_model():
    model = keras.models.Sequential()
    model.add(layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    
    #optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mse', optimizer='adam', metrics=['mae', 'mse'])
    
    return model

model = build_model()

model.summary()
```

#モデル訓練

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0:
            print('')
        print(',', end='')
        
EPOCHS = 500

history = model.fit(normed_train_data, train_labels, 
                   epochs=EPOCHS, validation_split=0.1, 
                   verbose=1)
```

```python
plt.plot(history.history['mae'])
plt.plot(history.history['val_mae'])
```

```python
plt.plot(history.history['mse'])
plt.plot(history.history['val_mse'])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,5])
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,20])
  plt.legend()
  plt.show()


plot_history(history)
```

```python
model = build_model()

# patience は改善が見られるかを監視するエポック数を表すパラメーター
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.1, verbose=1, callbacks=[early_stop])

plot_history(history)
```

```python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)
print("Testing set Mean Abs Error: {:5.2f} MPG", format(mae))
```

```python
test_predictions = model.predict(normed_test_data).flatten()
test_predictions
```

```python
plt.figure()
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Value [MPG]')
plt.ylabel('Predictions [MPG]')
plt.axis('equal')
plt.axis('square')
plt.plot([190, 250], [190, 250])
```

```python
error = test_predictions - test_labels
plt.hist(error, bins=100, histtype='step')
plt.xlabel("Prediction Error [MPG]")
plt.ylabel("Count")
plt.grid()
```
