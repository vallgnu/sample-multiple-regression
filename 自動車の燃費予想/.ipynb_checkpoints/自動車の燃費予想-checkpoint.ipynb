ライブラリのインポート

```python
import pathlib 

import matplotlib.pyplot as plt 
import pandas as pd 
import seaborn as sns

import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers

print(tf.__version__)
```

データセットのダウンロード

```python
dataset_path = keras.utils.get_file("auto-mpg.data", "https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data")
dataset_path
```

```python
column_names = ['MPG','Cylinders','Displacement','Horsepower','Weight',
                'Acceleration', 'Model Year', 'Origin'] 

raw_dataset = pd.read_csv(dataset_path, names=column_names, na_values="?", comment='\t', 
                         sep=" ", skipinitialspace=True)

dataset = raw_dataset.copy()
dataset.tail()
```

データクレンジング

```python
dataset.isna().sum()
```

```python
dataset = dataset.dropna()
```

```python
plt.plot(dataset['MPG'], label="MPG")
plt.plot(dataset['Cylinders'], label="Cylin")
plt.plot(dataset['Displacement'], label="Disp")
plt.legend()
```

```python
origin = dataset.pop('Origin')
origin
```

```python
dataset
```

```python
dataset['USA'] = (origin == 1) * 1
dataset['Europe'] = (origin == 2) * 1
dataset['Japan'] = (origin == 3) * 1
dataset.tail()
```

データセットを訓練用とテスト用に分割

```python
train_dataset = dataset.sample(frac=0.8, random_state=0)
test_dataset = dataset.drop(train_dataset.index)
```

データの観察

```python
sns.pairplot(train_dataset[["MPG", "Cylinders", "Displacement", "Weight"]], diag_kind="kde")
```

```python
train_stats = train_dataset.describe()
train_stats.pop("MPG")
train_stats = train_stats.transpose()
```

```python
train_stats
```

ラベルと特徴量の分離

```python
train_labels = train_dataset.pop("MPG")
test_labels = test_dataset.pop("MPG")
```

```python
train_labels
```

データの正規化

```python
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

normed_train_data = norm(train_dataset)
normed_test_data = norm(test_dataset)
```

```python
normed_train_data
```

```python
plt.scatter(normed_train_data['Cylinders'], normed_train_data['Displacement'])
```

モデルの構築

```python
def build_model():
    model = keras.models.Sequential()
    model.add(layers.Dense(64, activation='relu', input_shape=[len(train_dataset.keys())]))
    model.add(layers.Dense(64, activation='relu'))
    model.add(layers.Dense(1))
    
    optimizer = tf.keras.optimizers.RMSprop(0.001)
    
    model.compile(loss='mse',
                optimizer=optimizer,
                metrics=['mae', 'mse'])

    return model
    
model = build_model()
    
model.summary()
```

```python
example_batch = normed_train_data[:10]
example_result = model.predict(example_batch)
example_result
```

モデルの訓練

```python
class PrintDot(keras.callbacks.Callback):
    def on_epoch_end(self, epoch, logs):
        if epoch % 100 == 0: print('')
        print('.', end='')
        
EPOCHS = 1000

#verbose=1で学習時のデータ表示
history = model.fit(normed_train_data, train_labels, 
                    epochs=EPOCHS, validation_split=0.2, 
                   verbose=1)#, callbacks=[PrintDot()])
```

```python
plt.plot(history.history['mae'])
plt.plot(history.history['val_mae'])
```

```python
plt.plot(history.history['mse'])
plt.plot(history.history['val_mse'])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_history(history):
  hist = pd.DataFrame(history.history)
  hist['epoch'] = history.epoch

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Abs Error [MPG]')
  plt.plot(hist['epoch'], hist['mae'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mae'],
           label = 'Val Error')
  plt.ylim([0,5])
  plt.legend()

  plt.figure()
  plt.xlabel('Epoch')
  plt.ylabel('Mean Square Error [$MPG^2$]')
  plt.plot(hist['epoch'], hist['mse'],
           label='Train Error')
  plt.plot(hist['epoch'], hist['val_mse'],
           label = 'Val Error')
  plt.ylim([0,20])
  plt.legend()
  plt.show()


plot_history(history)
```

```python
model = build_model()

# patience は改善が見られるかを監視するエポック数を表すパラメーター
early_stop = keras.callbacks.EarlyStopping(monitor='val_loss', patience=10)

history = model.fit(normed_train_data, train_labels, epochs=EPOCHS,
                    validation_split = 0.2, verbose=1, callbacks=[early_stop])

plot_history(history)
```

```python
loss, mae, mse = model.evaluate(normed_test_data, test_labels, verbose=2)

print("Testing set Mean Abs Error: {:5.2f} MPG".format(mae))
```
